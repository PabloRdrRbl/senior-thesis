```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

The 2D convection is represented by the pair of coupled partial differential equations below:

$$
\dfrac{\partial u}{\partial t} + u \dfrac{\partial u}{\partial x} + v \dfrac{\partial u}{\partial y} = 0
$$


$$
\dfrac{\partial v}{\partial t} + u \dfrac{\partial u}{\partial x} + v \dfrac{\partial u}{\partial y} = 0
$$

Discretizing these equations, we end up with:

$$
\dfrac{u^{n+1}_{i,j}-u^n_{i,j}}{\Delta t} + u_{i,j}^{n} \dfrac{u_{i,j}^{n}-u_{i-1,j}^n}{\Delta x}+ v_{i,j}^{n} \dfrac{u_{i,j}^{n}-u_{i,j-1}^{n}}{\Delta y} = 0
$$


$$
\dfrac{v^{n+1}_{i,j}-v^n_{i,j}}{\Delta t} + u_{i,j}^{n} \dfrac{v_{i,j}^{n}-v_{i-1,j}^n}{\Delta x}+ v_{i,j}^{n} \dfrac{v_{i,j}^{n}-v_{i,j-1}^{n}}{\Delta y} = 0
$$

The unknown variables in these case are two $u_{i,j}^{n+1}$ and $v_{i,j}^{n+1}$. Solving these equations, it can also be seen that they are coupled:

$$
u_{i,j}^{n+1} = u_{i,j}^{n} - u_{i,j}^{n} \dfrac{\Delta t}{\Delta x} \left( u_{i,j}^{n}-u_{i-1,j}^n \right) - v_{i,j}^{n} \dfrac{\Delta t}{\Delta y} \left( u_{i,j}^{n}-u_{i,j-1}^n \right)
$$


$$
v_{i,j}^{n+1} = v_{i,j}^{n} - u_{i,j}^{n} \dfrac{\Delta t}{\Delta x} \left( v_{i,j}^{n}-v_{i-1,j}^n \right) - v_{i,j}^{n} \dfrac{\Delta t}{\Delta y} \left( v_{i,j}^{n}-v_{i,j-1}^n \right)
$$


The set of initial conditions will be the same as for the convection:

$$
u,v = \begin{Bmatrix} 2 & \text{ for} & 0.5 \leqslant x, y \leqslant 1 \\ 1 & \text{ for} & \text{everywhere else} \\ \end{Bmatrix}
$$

Doing the same for the boundary conditions:

$$
u=1, v=1 \text{  for  } \begin{Bmatrix} x=0,\ 2 \\ y=0,\ 2 \end{Bmatrix}
$$

```python
nx = 101
ny = 101
nt = 80
c = 1
dx = 2 / (nx - 1)
dy = 2 / (ny - 1)
CFL = 0.2
dt = CFL * dx / c

x = np.linspace(0,2,nx)
y = np.linspace(0,2,ny)

X, Y = np.meshgrid(x,y)

u = np.ones((ny,nx))
v = np.ones((ny,nx))

u[int(0.5/dy):int(1/dy+1), int(0.5/dx):int(1/dx+1)]=2
v[int(0.5/dy):int(1/dy+1), int(0.5/dx):int(1/dx+1)]=2
```

```python
#simple code to 3D plotting
fig = plt.figure(figsize=(11,7), dpi=100)
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, u, cmap=cm.viridis)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')
ax.set_title('U velocity field map')
ax.grid(True)
```

```python
#simple code to 3D plotting
fig = plt.figure(figsize=(11,7), dpi=100)
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, v, cmap=cm.viridis)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')
ax.set_title('V velocity field map')
ax.grid(True)
```

```python
un = np.ones((ny,nx))
vn = np.ones((ny,nx))

for n in range(nt+1):
    un = u.copy()
    vn = v.copy()
    u[1:,1:] = (un[1:,1:]-(un[1:,1:]*c*dt/dx*(un[1:,1:]-un[1:,:-1]))-vn[1:,1:]*c*dt/dy*(un[1:,1:]-un[:-1,1:]))
    v[1:,1:] = (vn[1:,1:]-(un[1:,1:]*c*dt/dx*(vn[1:,1:]-vn[1:,:-1]))-vn[1:,1:]*c*dt/dy*(vn[1:,1:]-vn[:-1,1:]))
    
    u[0, :] = 1
    u[-1, :] = 1
    u[:, 0] = 1
    u[:, -1] = 1
    
    v[0, :] = 1
    v[-1, :] = 1
    v[:, 0] = 1
    v[:, -1] = 1
```

```python
#simple code to 3D plotting
fig = plt.figure(figsize=(11,7), dpi=100)
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, u, cmap=cm.viridis)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')
ax.set_title('U velocity field map')
ax.grid(True)
```
