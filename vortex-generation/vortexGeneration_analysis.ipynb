# 2. Inlet perturbation BC to force oscillation vortex sheding

```python
y = np.linspace(-1,1,250)
t = np.linspace(0,100,250)

Y, T = np.meshgrid(y, t)

Z = 1+2*np.exp(-((T-20)**2)/(0.025)-((Y-1)**2)/(1.5))

fig = plt.figure(figsize=(11,7), dpi=100)
ax = fig.gca(projection='3d')
surf = ax.plot_surface(T, Y, Z, rstride = 1, cstride = 1, cmap = cm.viridis, linewidth = 0, antialiased = False)
ax.set_xlabel('$t$')
ax.set_ylabel('$y$')
ax.set_zlabel('$V$')
ax.set_zlim([0,3])
ax.set_title('Inlet Velocity BC')
```
