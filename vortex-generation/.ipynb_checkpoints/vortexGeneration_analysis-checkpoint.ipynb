# 1. Basic function declaration

```python
#Class to create an HTML table with some data - original from Yves Dubief
class ListTable(list):
    def _repr_html_(self):
        html = ["<table>"]
        for row in self:
            html.append("<tr>")
            for col in row:
                html.append("<td>{0}</td>".format(col))
            html.append("</tr>")
        html.append("</table>")
        return ''.join(html)
```

# 2. Inlet perturbation BC to force oscillation vortex sheding

The boundary condition applied on the inlet to perturbate the system is created by using function <tt> Z </tt>. It creates a small peak along y-axis in a very small step of time. It quickly goes to the original value again. The variables allow the movement of the perturbation in time and space, and also to modify the steady value.

```python
y = np.linspace(-1,1,250)
t = np.linspace(0,100,250)

Y, T = np.meshgrid(y, t)

steadyVal = 1
pertbTime = 20
peakAmp = 2
yPos = 1
timeCmp = 0.025
yCmp = 1.5

Z = steadyVal+peakAmp*np.exp(-((T-pertbTime)**2)/(timeCmp)-((Y-yPos)**2)/(yCmp))

fig = plt.figure(figsize=(11,7), dpi=100)
ax = fig.gca(projection='3d')
surf = ax.plot_surface(T, Y, Z, rstride = 1, cstride = 1, cmap = cm.viridis, linewidth = 0, antialiased = False)
ax.set_xlabel('$t$')
ax.set_ylabel('$y$')
ax.set_zlabel('$V$')
ax.set_zlim([0,3])
ax.set_title('Inlet Velocity BC')
```

## 3.5. Wave interpolation and calculations

It can be seen that all different cuts along the X-axis have the oscillations coupled in the above and below streamline. After $t=100\ s$ the oscillations are constant and repetitive, so in order to analyze them, a curve fitting will be done - amplitude, frequency, phase and offset will be computed for the different curves. The original idea from the fitting was taken from: http://scipy-cookbook.readthedocs.io/items/FittingData.html

```python
#Restrict the data of the main streamline from t=100s onwards
above100 = above[100:,:,0]
below100 = below[100:,:,0]

#Creation of time values
time = np.linspace(100, 200, 100)
```

```python
def sine_fit(p, x): #Classical sine wave function
    return p[0]*np.cos(2*np.pi/p[1]*x+p[2]) + p[3]

def dist_func(p, x, y): #Distance to the sine function
     return sine_fit(p, x) - y 

#Allocation of space for (amplitude, frequency, phase and offset) for the whole size of above100 and below100
p1a = np.zeros((4,np.shape(above100)[1]))
p1b = np.zeros((4,np.shape(below100)[1]))

#Go over above100 to compute the wave-parameters of each x-axis cut
for i in range(np.shape(above100)[1]):
    press = above100[:,i]
    #The intial guess for each case is composed of:
    #   Amplitude: sqrt(2)*standardDeviation(data) is one way of having a guess on amplitude (http://www.dspguide.com/ch2/2.htm)
    #   Frequency: let's start with a value high enough that will be afterwards corrected
    #   Phase: let's assume zero for the guess
    #   Offset: mean(data) gives the mean value of the wave function
    guess = [np.sqrt(2)*np.std(press), 11, 0., np.mean(press)]
    #Calculation of the optimum parameters for the data set
    p1a[:,i], success = scipy.optimize.leastsq(dist_func, guess[:], args=(time, press))

#Go over below100 to compute the wave-parameters of each x-axis cut (following the above procedure)
for i in range(np.shape(below100)[1]):
    press = below100[:,i]
    guess = [np.sqrt(2)*np.std(press), 11, 0., np.mean(press)]
    p1b[:,i], success = scipy.optimize.leastsq(dist_func, guess[:], args=(time, press))
```

```python
#The fitting obtained is shown for the same x-axis cuts done before
fig, ax = plt.subplots(nrows=6, ncols=2, sharex=True, figsize=(17,20), dpi=100)

plt.subplot(6, 2, 1)
plt.plot(np.linspace(0, timesteps, timesteps), above[:,np.abs(above[0,:,5]-2).argmin(),0], 'r.', label='x = 2')
plt.plot(np.linspace(100,200,100),sine_fit(p1a[:,np.abs(above[0,:,5]-2).argmin()], np.linspace(100,200,100)), 'k-', label='Curve-fit')
plt.xlim([100,200])
plt.title('Above streamline')
plt.legend()

plt.subplot(6, 2, 2)
plt.plot(np.linspace(0, timesteps, timesteps), below[:,np.abs(below[0,:,5]-2).argmin(),0], 'r.', label='x = 2')
plt.plot(np.linspace(100,200,100),sine_fit(p1b[:,np.abs(below[0,:,5]-2).argmin()], np.linspace(100,200,100)), 'k-', label='Curve-fit')
plt.xlim([100,200])
plt.title('Below streamline')
plt.legend()

xCuttingPoint = np.array([4,6,8,10,12])
color = ['g','b','y','k','c']

for i in range(0,5):   
    plt.subplot(6, 2, 3+2*i)
    plt.plot(np.linspace(0, timesteps, timesteps), above[:,np.abs(above[0,:,5]-xCuttingPoint[i]).argmin(),0], '.', color=color[i], label='x = %i' %xCuttingPoint[i])
    plt.plot(np.linspace(100,200,100),sine_fit(p1a[:,np.abs(above[0,:,5]-xCuttingPoint[i]).argmin()], np.linspace(100,200,100)), 'k-', label='Curve-fit')
    plt.xlim([100,200])
    plt.ylabel('Pressure')
    plt.xlabel('Line length')
    plt.legend()

    plt.subplot(6, 2, 4+2*i)
    plt.plot(np.linspace(0, timesteps, timesteps), below[:,np.abs(below[0,:,5]-xCuttingPoint[i]).argmin(),0], '.', color=color[i], label='x = %i' %xCuttingPoint[i])
    plt.plot(np.linspace(100,200,100),sine_fit(p1b[:,np.abs(below[0,:,5]-xCuttingPoint[i]).argmin()], np.linspace(100,200,100)), 'k-', label='Curve-fit')
    plt.xlim([100,200])
    plt.ylabel('Pressure')
    plt.xlabel('Line length')
    plt.legend()
```

## 3.6. Active flow control - rotational BC - calculations

One of the most classic methods to AFC is rotation of the cylinder. Although this won't be very practical for an airfoil, in the case of a cylinder can  be more easily achieved. In order to achieve a good understanding, a time dependant function will be created: a step fuction is needed to turn ON/OFF the AFC system and a sine or cosine function will be needed to change the direction of rotation. These gunction can be written by using the Heaviside function as follow:

$$
g(t) = u_c(t)\ f(t-c)
$$

where $u_c(t)$ is the Heaviside function and $f(t-c)$ is the function that the system will take when it is ON. Applying th Laplace transform to the equation, it yields:

$$
\mathcal{L} (u_c(t)\ f(t-c)) = \int_{0}^{\infty} e^{-st}\ u_c(t)\ f(t-c)\ dt = \int_{c}^{\infty} e^{-st}\  f(t-c)\ dt  
$$

Applying a change in variable inside the integral (replacing $u = t-c$):
$$
\mathcal{L} (u_c(t)\ f(t-c)) = \int_{0}^{\infty} e^{-s(u+c)}\ f(u)\ du = \int_{0}^{\infty} e^{-su}\ e^{-sc}\ f(u)\ du   
$$

Integrating:
$$
\mathcal{L} (u_c(t)\ f(t-c)) = e^{-sc}\ \int_{0}^{\infty} e^{-su}\ f(u)\ du = e^{-cs} F(s) 
$$

Taking the inverse Laplace transform, the function in time domain is obtained:
$$
\mathcal{L}^{1}(e^{-cs} F(s)) = u_c(t)\ f(t-c)
$$

In this case, the function $F(s) = \mathcal{L}(sin(\omega t))=\dfrac{\omega}{s^2+\omega^2}$

```python
s, t = sp.symbols('s, t')
w = sp.symbols('w', real = True)
```

```python
switchTime = 75
rotatingBC_exp = sp.exp(-switchTime*s)*w/(s**2+w**2)

rotatingBC = sp.inverse_laplace_transform(rotatingBC_exp, s, t)
```

Although simulation has a time limit of 200 s, the time will only go until 150 s to see if the vortex shedding is eliminated if the AFC is turned of or if the AFC is still necessary. This will made thanks to the <tt> .csv </tt> boundary condition introduction by simply limiting the time to 150 s - when OpenFOAM reads the file will clamp the values when time goes above 150 s, the value of the rotating velocity will be the last value of the file - which is indeed zero.


```python
rotBCtime = np.concatenate([np.linspace(0,74.9,3),np.linspace(75.1,150,1007)])

rotatingBCvals = np.zeros(len(rotBCtime))
for i in range(len(rotBCtime)):
    rotatingBCvals[i] = rotatingBC.evalf(subs={t:rotBCtime[i], w:np.mean([np.mean(p1a[1,:]),np.mean(p1b[1,:])])/(2*np.pi)})
    
fig, ax = plt.subplots(1, figsize=(20,4), dpi = 100)
ax.plot(rotBCtime, rotatingBCvals)
ax.set_xlabel('Time')
ax.set_ylabel('Rotating wall velocity')
```

The BC of <tt> Rotating wall </tt> receives a list of points with the velocity value in each case instead a function of time $f(t)$. So let's save the computed points in a <tt> .csv </tt> file to use it as the simulation boundary condition.

```python
dumpMatrix = np.vstack((rotBCtime, rotatingBCvals)).T
np.savetxt("rotatingBC.csv", dumpMatrix, delimiter=",")
```

## 3.6. Active flow control - momentum insertion BC - calculations

At first, only flow momentum introduction will be considered. If the sine has negative values, will have suction. In order to achieve just introduction, a displacement of the sine function must be performed, applying an offset in the vertical direction. If instead one impulse, two are applied - one in each direction - a rectangular function can be obtained. Given that <tt> Heaviside </tt> is not implemented in OpenFOAM, it will be represented with $tanh(x)$. Multiplying the obtainted function times the offsetSine will yield the final momentum insertion BC.

```python
onTime = 75
offTime = 150

rectangular_exp = -(sp.exp(-offTime*s)-sp.exp(-onTime*s))/s
rectangular = sp.inverse_laplace_transform(rectangular_exp, s, t)

rectangularHeav = list(str(rectangular))
rectangularTanh = []

i = 0

while i < len(str(rectangular)):
    if rectangularHeav[i] == 'H':
        rectangularTanh.append('t')
        rectangularTanh.append('a')
        rectangularTanh.append('n')
        rectangularTanh.append('h')
        i += len('Heaviside')
    else:
        rectangularTanh.append(str(rectangular)[i])
        i += 1

rectangularTanh = sp.parsing.sympy_parser.parse_expr(''.join(rectangularTanh))
```

```python
offsetSine = sp.sin(t) + 1

momBC = rectangularTanh*offsetSine

print(momBC)
```

```python
momBCtime = np.concatenate([np.linspace(0,69.9,3),np.linspace(71.1,154.9,1007),np.linspace(155.1,200,3)])

momentumBCvals = np.zeros(len(momBCtime))
for i in range(len(momBCtime)):
    momentumBCvals[i] = momBC.evalf(subs={t:momBCtime[i]})
    
fig, ax = plt.subplots(1, figsize=(20,4), dpi = 100)
ax.plot(momBCtime, momentumBCvals)
ax.set_xlabel('Time')
ax.set_ylabel('Flow velocity')
```

## 3.7. AFC - momentum blowing and suction (zero $ \overline{m v}$ introduction) - calculations

Considering the same function as before but not offsetting the sine to have both blowing and suction in the boundary condition

```python
onTime = 75
offTime = 150

rectangular_exp = -(sp.exp(-offTime*s)-sp.exp(-onTime*s))/s
rectangular = sp.inverse_laplace_transform(rectangular_exp, s, t)

rectangularHeav = list(str(rectangular))
rectangularTanh = []

i = 0

while i < len(str(rectangular)):
    if rectangularHeav[i] == 'H':
        rectangularTanh.append('t')
        rectangularTanh.append('a')
        rectangularTanh.append('n')
        rectangularTanh.append('h')
        i += len('Heaviside')
    else:
        rectangularTanh.append(str(rectangular)[i])
        i += 1

rectangularTanh = sp.parsing.sympy_parser.parse_expr(''.join(rectangularTanh))
```

```python
notOffsetSine = sp.sin(t)

BSmomBC = rectangularTanh*notOffsetSine

print(BSmomBC)
```

```python
BSmomBCtime = np.concatenate([np.linspace(0,69.9,3),np.linspace(71.1,154.9,1007),np.linspace(155.1,200,3)])

BSmomBCvals = np.zeros(len(BSmomBCtime))
for i in range(len(BSmomBCtime)):
    BSmomBCvals[i] = BSmomBC.evalf(subs={t:BSmomBCtime[i]})
    
fig, ax = plt.subplots(1, figsize=(20,4), dpi = 100)
ax.plot(BSmomBCtime, BSmomBCvals)
ax.set_xlabel('Time')
ax.set_ylabel('Flow velocity')
```
