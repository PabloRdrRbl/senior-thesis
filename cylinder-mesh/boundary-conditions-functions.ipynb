```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
def plot3d(X, Y, Z,title):
    fig = plt.figure(figsize=(11,7), dpi=100)
    ax = fig.gca(projection='3d')
    surf = ax.plot_surface(X, Y, Z, rstride = 1, cstride = 1, cmap = cm.viridis, linewidth = 0, antialiased = False)
    ax.set_xlabel('$t$')
    ax.set_ylabel('$y$')
    ax.set_zlabel('$V$')
    ax.set_zlim([0,3])
    ax.set_title('%s' %title)
```

```python
def plot2d(X, Y, title):
    fig = plt.figure(figsize=(7,5), dpi=100)
    ax = fig.gca()
    ax.plot(X, Y)
    ax.set_xlabel('$t$')
    ax.set_ylabel('$V$')
    ax.set_title('%s' %title)
```

```python
y = np.linspace(-1,1,250)
t = np.linspace(0,100,250)

Y, T = np.meshgrid(y, t)

Z = 1+2*np.exp(-((T-20)**2)/(0.025)-((Y-1)**2)/(1.5))

plot3d(T, Y, Z, 'Inlet Velocity BC')
```

```python
peaks = np.array([34.5019,40.0039,45.4977,51.0048,56.5053,62.0035,67.5011,72.9983,78.4951,84.0032,89.4989,95.0044])

increment = peaks[1:] - peaks[0:-1]

vst = peaks[0]                        #vortex shedding time
frequency = np.mean(1/increment)      #vortex shedding frequency

AFCv = (1+np.tanh(((t)+t*np.sin(2*np.pi*frequency*t))-2*vst))*np.cos(45)

plot2d(t, AFCv, 'AFC velocity BC')
```
